```python
import pandas as pd
import matplotlib.pyplot as plt


dataset_name = "company_sales_data.csv"
data = pd.read_csv(dataset_name)


face_sales_data = data[['month_number', 'facecream', 'facewash']]


face_sales_data.plot(x='month_number', kind='bar', figsize=(10, 6))
plt.title('Face Cream and Facewash Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['total_profit'].plot(kind='hist', bins=5, title='Total Profit Histogram')
plt.xlabel('Total Profit Range')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)

df
```

```python
profitList = df ['total_profit'].tolist()
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data')
plt.show()
```

```python
import matplotlib.pyplot as plt

monthList  = df['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()

plt.bar([a-0.25 for a in monthList], faceCremSalesData, width= 0.25)
plt.bar([a+0.25 for a in monthList], faceWashSalesData, width= -0.25)


plt.xticks(monthList)
plt.xlabel('month list')
plt.ylabel('total sales')
plt.show()
```
